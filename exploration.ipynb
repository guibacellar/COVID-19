```python
%matplotlib widget

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/all_timeline_with_features.csv')
```

```python
df.columns
```

```python
COUNTRY_CODE = 'BR'

target_country = df[(df['countrycode'] == COUNTRY_CODE) & (df['totalcases'].notna())]
target_country = target_country[:-1]
target_country.head(100)
```

```python
# Cases VS Deaths
fig, ax1 = plt.subplots(figsize=(9, 7),constrained_layout=True)  # Set Figure Size
plt.xticks(rotation=90)  # Rotate X Label

fig.suptitle(f"[{COUNTRY_CODE}] - Total Cases VS Total Death's")

# Total Cases
ax1.plot(
    target_country['date'],
    target_country['totalcases'],
    color='blue',
    marker='d'
)
ax1.set_xlabel('Dates')  
ax1.set_ylabel("Infection Cases", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')


# TOtal Deaths
ax2 = ax1.twinx()
ax2.plot(
    target_country['date'],
    target_country['totaldeaths'],
    color='red',
    marker='+',
    label="Death's"
)
ax2.set_ylabel("Death's", color='red')
ax2.tick_params(axis='y', labelcolor='red')


# Fig2 - Evolution Rate Day
fig2, ax3 = plt.subplots(figsize=(9, 3),constrained_layout=True)  # Set Figure Size
plt.xticks(rotation=90)  # Rotate X Label

fig2.suptitle(f"[{COUNTRY_CODE}] - Death Rate per Day (%)")
ax3.plot(
    target_country['date'],
    target_country['ft13_death_percent'],
    color='red',
    marker='o',
    label="Rate"
)
ax3.set_ylabel("Rate", color='red')
ax3.tick_params(axis='y', labelcolor='red')
```

```python
# New Cases per Day
fig, ax1 = plt.subplots(figsize=(9, 7),constrained_layout=True)  # Set Figure Size
plt.xticks(rotation=90)  # Rotate X Label

fig.suptitle(f"[{COUNTRY_CODE}] - Total Cases VS New Cases per Day")

# Total Cases
ax1.plot(
    target_country['date'],
    target_country['totalcases'],
    color='blue',
    marker='d'
)
ax1.set_xlabel('Dates')  
ax1.set_ylabel("Infection Cases", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')


# New Cases per Day
ax2 = ax1.twinx()
ax2.scatter(
    target_country['date'],
    target_country['ft04_new_cases_per_day'],
    color='red',
    marker='P'
)
ax2.set_ylabel("New Cases per Day", color='red')
ax2.tick_params(axis='y', labelcolor='red')


# Fig2 - Death Rate per Day
fig2, ax3 = plt.subplots(figsize=(9, 3),constrained_layout=True)  # Set Figure Size
plt.xticks(rotation=90)  # Rotate X Label

fig2.suptitle(f"[{COUNTRY_CODE}] - Cases Evolution Rate")
ax3.plot(
    target_country['date'],
    target_country['ft10_cases_evolution_rate'],
    color='red',
    marker='+'
)
ax3.set_ylabel("Rate", color='red')
ax3.tick_params(axis='y', labelcolor='red')
```

```python
# Cases Vs Case Evolution Rate
fig, ax1 = plt.subplots(figsize=(9, 7),constrained_layout=True)  # Set Figure Size
plt.xticks(rotation=90)  # Rotate X Label

fig.suptitle(f"[{COUNTRY_CODE}] - Total Deaths VS Deaths per Day")

# Evolution Rate 
ax1.plot(
    target_country['date'],
    target_country['totaldeaths'],
    color='red',
    marker='+'
)
ax1.set_ylabel("Total Deaths", color='red')
ax1.tick_params(axis='y', labelcolor='red')


# Total Cases
ax2 = ax1.twinx()
ax2.scatter(
    target_country['date'],
    target_country['ft05_new_deaths_per_day'],
    color='blue',
    marker='d'
)
ax2.set_xlabel('Dates')  
ax2.set_ylabel("Deaths per Day", color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
```
